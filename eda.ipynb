### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from collections import defaultdict
warnings.filterwarnings("ignore")
```

### Read datasets

```python
data = pd.read_csv("data.csv")
```

### Visualizations
#### Histogram

```python
# Create bins for histogram
income = data['annual_income']
income_dict = defaultdict(int)
for i in income.values:
    if i <= 25000:
        income_dict[25] += 1
    elif i > 25000 and i <= 50000:
        income_dict[50] += 1
    elif i > 50000 and i <= 75000:
        income_dict[75] += 1
    elif i > 75000 and i <= 100000:
        income_dict[100] += 1
    elif i > 10000 and i <= 125000:
        income_dict[125] += 1
    elif i > 125000 and i <= 150000:
        income_dict[150] += 1
    elif i > 150000 and i <= 175000:
        income_dict[175] += 1
    elif i > 175000 and i <= 200000:
        income_dict[200] += 1
    elif i > 200000 and i <= 225000:
        income_dict[225] += 1
    elif i > 225000 and i <= 250000:
        income_dict[250] += 1
    elif i > 250000 and i <= 275000:
        income_dict[275] += 1
    elif i > 275000 and i <= 300000:
        income_dict[300] += 1
    elif i > 300000 and i <= 325000:
        income_dict[325] += 1
    elif i > 325000 and i <= 350000:
        income_dict[350] += 1
    elif i > 350000 and i <= 375000:
        income_dict[375] += 1
    elif i > 375000 and i <= 400000:
        income_dict[400] += 1
    elif i > 400000 and i <= 425000:
        income_dict[425] += 1
    elif i > 425000 and i <= 450000:
        income_dict[450] += 1
    else:
        income_dict[475] += 1

# Create plot
fig, ax = plt.subplots(figsize=(10,7))

# Sort dictionary
t = sorted(income_dict.items(), key=lambda x: x[0])

# Distribute keys and values into lists
cat = []
count = []
for b in t:
 cat.append(b[0])
 count.append(b[1])

# Create histogram
ax.bar(x = cat, height = count, width=25, edgecolor="white")
ax.set_xticks(ticks=[0, 62.5, 112.5, 162.5, 212.5, 262.5, 312.5, 362.5, 412.5, 462.5],
              labels=[0, 75, 125, 175, 225, 275, 325, 375, 425, "475+"])
ax.set_xticklabels([0, 75, 125, 175, 225, 275, 325, 375, 425, "475+"], fontsize=14)
ax.set_yticks([0, 1000, 2000, 3000, 4000, 5000, 6000],)
ax.set_yticklabels([0, 1000, 2000, 3000, 4000, 5000, 6000], fontsize=14)
plt.tick_params(axis='x', which='both', pad=5)
ax.set_title("Distribution of Client Income", fontsize=24, pad=20)
ax.set_xlabel("Income (in thousands)", fontsize=18)
ax.set_ylabel("Count", fontsize=18)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['top'].set_visible(False)
plt.show()
```

#### Bar plot

```python
# Create plot
fig, ax = plt.subplots(figsize=(10,7))

# Compute number of no defaults and defaults
default = data['default']
default_dict = defaultdict(int)
for d in default.values:
    if d == 0:
        default_dict['no default'] += 1
    elif d == 1:
        default_dict['default'] += 1

# Create bar plot
bars = ax.bar(x = list(default_dict.keys()), height = list(default_dict.values()), 
              width=0.8, edgecolor="white", color=['#1f77b4', '#d62728'])
ax.bar_label(bars,fontsize=20)
ax.set_xticklabels(labels = list(default_dict.keys()), fontsize=14)
ax.set_yticklabels([0, 5000, 10000, 15000, 20000, 25000, 30000], fontsize=14)
ax.set_title("Distribution of Client Defaults", fontsize=24, pad=20)
ax.set_xlabel("Client Default Status", fontsize=18)
ax.set_ylabel("Count", fontsize=18)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['top'].set_visible(False)
plt.show()
```
